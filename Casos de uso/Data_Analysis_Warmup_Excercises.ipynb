## Apply Method

```python
# Creamos una función
def marks(x):
  return x/2
```

```python
# Es una forma de aplicar una función al data frame
df1['Marks'].apply(marks)
```

```python
# Lo añadimos como nueva columna
df1['Half_marks'] = df1['Marks'].apply(marks)
df1
```

```python
# Otra forma de aplicar la función
df1['Marks'].apply(lambda x: x/2)
```

## Map function

```python
# Cambiamos los valores de una columna de acuerdo a un patrón
df1['Male_female'] = df1['Gender'].map({'Male': 1, 'Female': 0})
df1
```

## Drop the new 2 columns

```python
df1.drop(['Half_marks', 'Male_female'], axis = 1, inplace = True)
df1
```
