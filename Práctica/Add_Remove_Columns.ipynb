# Add column or columns

```python
import pandas as pd
dict1 ={'Name':['Rafa','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
               }
df1=pd.DataFrame(dict1)
df1
```

## Add column using a list

```python
# Creamos la lista
Grades = ['AA','AB','AA','AB','AC','BA','BB']
```

```python
# La añadimos al dataframe
df1['Grades'] = Grades
df1
```

```python
# Borramos la columna
df1.drop('Grades', axis = 1, inplace = True)
# df1.drop(['Marks','Grades'], axis = 1, inplace = True)
df1
```

## Add column using a pandas series

```python
# Creamos la serie
s1 = pd.Series(['AA','AB','AA','AB','AC','BA','BB'])
```

```python
# Añadimos la nueva columna
df1['Grades'] = s1
```

```python
# Borramos la columna
df1.drop('Grades', axis = 1, inplace = True)
```

## Add column using dictionary

```python
# Creamos el diccionario
d={'Anand':'Rafa','Ahmedabad':'Addhya',
   'Nadiad':'Krisha','Baroda':'Vedant',
   'khambhat':'Parshv',
   'Khambhat':'Mittal','Ahmedabad1':'Archana'}
d
```

```python
# Añadimos la columna
df1['City'] = d
df1
```

```python
# Borramos la columna
df1.drop('City', axis = 1, inplace = True)
```

## Insert a column in a particular position

```python
# Creamos la serie
s1 = pd.Series(['AA','AB','AA','AB','AC','BA','BB'])
```

```python
# Insertamos la columna en la primera posición (0)
df1.insert(0, 'Grades', s1)
df1
```

```python
# Borramos la columna
df1.drop('Grades', axis = 1, inplace = True)
df1
```

## Adding column from existing columns

```python
# Creamos una columna usando columnas que ya existen en un dataframe
df1['Name_Grades'] = df1['Name'] + " " + df1['Grades']
df1
```

```python
# Borramos la columna
df1.drop('Grades', axis = 1, inplace = True)
df1
```

## Using assing

```python
# El método assing creará un nuevo dataframe con la nueva columna
df2 = df1.assign(Grades=['AA','AB','AA','AB','AC','BA','BB'])
df2
```

```python
# También podemos añadir más columnas a la vez
# Creamos una serie para añadirla
s2 = pd.Series([11, 22, 33, 44, 55, 66, 77])
df3 = df1.assign(Grades=['AA','AB','AA','AB','AC','BA','BB'], temp = s2)
df3
```

```python
# Podemos añadir la columna de las calificaciones más 1
df4 = df1.assign(marks1 = lambda x:x.Marks + 1)
df4
```

```python
# También se podría de esta manera
df4 = df1.assign(marks2 = df1['Marks']+ 1)
df4
```
