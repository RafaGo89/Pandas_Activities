## Add and remove rows

```python
import pandas as pd
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
               }
df1=pd.DataFrame(dict1)
df1
```

## Add rows using concat

```python
# Creamos la nueva fila
df2 = pd.DataFrame({'Name': 'Rafael', 'Marks': [99]})
```

```python
# La concatenamos al dataframe
# ignore_idex = True es para que se siga con la cuenta del index existente
df3 = pd.concat([df1, df2], ignore_index = True)
df3

# Para agregarlo al incio del data frame solo se invierte
# df3 = pd.concat([df2,df1])
```

## Add row using iloc

```python
# Añadimos la nueva fila en una posición dada
df1.iloc[1] = ['Anil', 89]
df1

# Si usamos una posición que ya tenga una fila, está se reemplazara
```

## Add row using loc

```python
# Añadimos una fila usando la etiqueta como posición
df1.loc[3] = ['Rafael', 90]
df1

# Si usamos una posición que ya tenga una fila, está se reemplazara
```

# Drop row(s)

```python
# Eliminamos la fila por el índice (eliminamos fila indice 0)
df1.drop([0], inplace = True)
df1
```

```python
# Se pueden eliminar más de 1 fila a la vez (eliminamos fila índice 1 y 2)
df1.drop([1, 2], inplace = True)
df1
```

```python
# Para reiniciar el conteo del indice
df1 = df1.reset_index(drop = True)
df1
```

## Removing rows per given condition

```python
# Volvemos a cargar el dataframe
df1=pd.DataFrame(dict1)
df1
```

```python
# Eliminar las filas donde la calificación se menor que 90
# Pasamos la condición a una variable
x = df1[df1['Marks']<90]
# Ejecutamos el método drop accediendo a los índices de la condición
df1 = df1.drop(x.index)
df1
```

```python
# También se pueden eliminar de esta manera
df1.drop(df1[df1['Marks']<90].index)
```
