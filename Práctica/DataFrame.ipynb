# DataFrames

## They can be created based on
### List
### Dictionary
### Numpy array
### Series

```python
import pandas as pd
```

```python
# Creación de dataframe simple
df = pd.DataFrame([11, 22, 33, 44, 55])
df
```

# Dataframes (list)

```python
list1 = [11, 22, 33, 44, 55]
df = pd.DataFrame(list1)

df

# df = pd.DataFrame([11, 22, 33, 44, 55])   Método directo
```

# Dataframes (nested list)

```python
list1 = [['Rafa', 21], ['Natalia', 20], ['Pedro', 30]]
df = pd.DataFrame(list1)
df
# df = pd.DataFrame([['Rafa', 21], ['Natalia', 20], ['Pedro', 30]])   Método directo
```

# Dataframes (dictionary)

```python
d = {'Name': ['Rafa', 'Natalia', 'Pedro'], 'Age': [21, 20, 30]}

df = pd.DataFrame(d)
df
# df = pd.DataFrame({'Name': ['Rafa', 'Natalia', 'Pedro'], 'Age': [21, 20, 30]})   Método directo
```

# Dataframes (combinations of series)

```python
d = {'Name': pd.Series(['Rafa', 'Natalia', 'Pedro']), 'Age': pd.Series([21, 20, 30])}

df = pd.DataFrame(d)
df
# df = pd.DataFrame({'Name': pd.Series(['Rafa', 'Natalia', 'Pedro']), 'Age': pd.Series([21, 20, 30])})   Método directo
```

# Customize Columns and index

```python
list1 = [['Rafa', 21], ['Natalia', 20], ['Pedro', 30]]

# Agregamo columns e index para establecerlos como queramos
df = pd.DataFrame(list1, columns = ['Name', "Age"], index = ['stu1', 'stu2', 'stu3'])
df
```

# Ejemplo

```python
df = pd.DataFrame([['Ahmedabad', 40.1], ['Anad', 35.2], ['Baroda', 38.5]],
                  columns = ["City", 'Temp'],
                  index = ["a", 2, 3])
df
```
