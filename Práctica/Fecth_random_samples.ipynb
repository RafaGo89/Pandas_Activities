## Sample
**sample ()** :
Returns a random sample of items from an axis of object.
We can apply this method to pandas series and pandas dataframes.

##Parameters by default:
n=None,

frac=None,

replace=False,

weights=None,

random_state=None,

axis=None,

##n:
int, optional
Number of items from axis to return. Cannot be used with 'frac'.

Default = 1 if 'frac' = None.

##frac:
float, optional

Fraction of axis items to return. Cannot be used with 'n'.

##replace:
bool, default False

Allow or disallow sampling of the same row more than once.

##weights:
str or ndarray-like, optional

Default 'None' results in equal probability weighting.

If passed a Series will align with target object on index.

##random_state :
int, array-like, BitGenerator, np.random.RandomState, optional

##axis :
{0 or ‘index’, 1 or ‘columns’, None},
 default None

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/iris_code.csv')
data.head()
```

```python
data.info()
```

```python
data['Species'].value_counts()
```

##Fetch 5 random samples from the dataset

```python
data.sample(n = 5)
```

##Fetch 50% of data from the dataset

```python
# frac devuelve una fracción de muestra, un porcentaje del dataset
data.sample(frac = 0.5)

# NOTA. no se puede usar 'frac' con 'n'
```

## random_state

```python
# Si pasamos algún valor 'x' a random_state, este hará que cada vez que
#ejecutemos el script, obtendremos la misma muestra
data.sample(n = 3, random_state = 10)
```

##Replace

```python
dict1 ={'Name':['Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[89,99,87,90,83,82],
               }
df1=pd.DataFrame(dict1)
df1
```

```python
# Si replace es True, podemos obtener la misma fila más de una vez
# en nuestra muestra
df1.sample(n = 3, replace = True )

# NOTA: por default es False
```

##Axis

```python
# Si axis es 0, obtenemos muestras de las fila
data.sample(n = 2, axis = 0)
```

```python
# Si axis es 1, obtenemos muestras de las columnas
data.sample(n = 2, axis = 1)
```

```python
# Si axis es 1, obtenemos muestras de las columnas
data.sample(n = 2, axis = 1)
```

##weights

```python
# Con weight, creamos una lista de la misma longitud que la cantidad
# de filas que tenemos. Cada valor es una probabilidad que se le asigna
# a cada fila, para hacer que algunas sean más o menos probable que
#aparezcan al tomar una muestra
sesgo = [0.1, 0.5, 0.8, 0.3, 0.2, 0.1]
df1.sample(n = 3, weights = sesgo)
```

##Pandas series

```python
# Podemos aplicar 'sample' de igual manera a una serie de pandas
df1['Name'].sample(n = 3)
```
