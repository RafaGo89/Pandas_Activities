#Group by
This function is used to **split data** into groups **based** on some **criteria**. So we can perform analytics **Groupwise**

```python
import pandas as pd
```

```python
d = {'City': ['San Luis Potosí','Buenos Aires','Managua',
              'San Luis Potosí','Buenos Aires','Managua'],
    'Temp':[40,35,32,45,37,34],
    'Humidity':[32,34,23,24,27,32]}
data = pd.DataFrame(d)
data
```

## Display average temperature

```python
from os import device_encoding
round(data['Temp'].mean(), 2)
```

## Display average temperature Citywsie

```python
grp = data.groupby('City')['Temp'].mean()
grp
```

## Display all the groups (iterate over the groups)

```python
grp = data.groupby('City')

for key, value in grp:
  print(key)
  print(value)
```

## Display the indices in each group

```python
grp = data.groupby('City')
grp.groups
```

##Display the first entry in all the groups formed

```python
grp = data.groupby('City')

grp.first()
```

```python
# Así se muestra la última entrada por grupo
grp.last()
```

##Count the number of entries in each group

```python
grp = data.groupby('City')

grp.size()
```

##Display a specific group out of all the groups

```python
grp = data.groupby('City')

grp.get_group('Buenos Aires')
```

##Apply functions to groupby groups (like min, max, mean, etc)

```python
grp = data.groupby('City')

grp.min()
```

```python
grp.max()
```

```python
grp.mean()
```

##Display average temperature CityWise with prefix Average_Temp_

```python
grp = data.groupby('City')['Temp'] \
.mean().add_prefix('Average_Temp_')
grp
```

## Display descriptive statistics GroupWise

```python
grp = data.groupby('City').describe()
grp
```

##Perform statistical computations using agg() function

```python
grp = data.groupby('City').agg(['min','max','mean'])
grp
```

## Create a function to get the statistics Groupwise

```python
def funcion(grp):
  return {'Min':grp.min(), 'Mean':grp.mean()}
```

```python
grp = data.groupby('City')['Temp'].apply(funcion)
grp
```

##Grouping by 2 columns

```python
# Se agrupa por ciudad, y dentro de cada ciudad por cada temperatura
grp = data.groupby(['City', 'Temp'])

for key, value in grp:
  print(key)
  print(value)
```
