#Group by without group by

## Operations to use Group by
**Split --> Apply --> Combine**

```python
import pandas as pd
```

```python
d = {'City': ['San Luis Potosí','Buenos Aires','Managua','San Luis Potosí',
              'Buenos Aires','Managua'],
    'Temp':[40,35,32,45,37,34]}

data = pd.DataFrame(d)
data
```

## Split

```python
# Creamos el primer grupo dividido por ciudad
grp_a = data[data['City'] == 'San Luis Potosí']
grp_a
```

```python
# Creamos el segundo grupo dividido por ciudad
grp_b = data[data['City'] == 'Buenos Aires']
grp_b
```

```python
# Creamos el tercer grupo dividido por ciudad
grp_c = data[data['City'] == 'Managua']
grp_c
```

## Apply

```python
# Aplicamos el promedio de temperatura al primer grupo
avg_temp_a = grp_a['Temp'].mean()
avg_temp_a
```

```python
# Aplicamos el promedio de temperatura al segundo grupo
avg_temp_b = grp_b['Temp'].mean()
avg_temp_b
```

```python
# Aplicamos el promedio de temperatura al tercer grupo
avg_temp_c = grp_c['Temp'].mean()
avg_temp_c
```

##Combine

```python
# Creamos un nuevo dataframe uniendo las ciudades con sus respectivos valores
# de temperatura promedio
dataf = pd.DataFrame({'City':['San Luis Potosí', 'Buenos Aires', 'Managua'],
                      'Temp':[avg_temp_a, avg_temp_b, avg_temp_c]})
dataf
```
