```python
import pandas as pd
```

### Customized Indexes

```python
s1 = pd.Series([11, 22, 33, 44, 55], index = ['a', 'b', 'c', 'd', 'e']) # Agregamos un Ã­ndice personalizado
s1
```

### Datatype of serie

```python
s1 = pd.Series([11, 22, 33, 44, 55], dtype = 'float')
s1
```

### Accesing data from the serie

```python
s1 = pd.Series([11, 22, 33, 44, 55])
```

```python
s1[0] # Accedemos al valor por su indice
```

```python
s1[1]
```

```python
s1[0:2]  # primer indice, desde donde; segundo indice, hasta cual menos 1
```

```python
s1[1:3]
```

### Operations on pandas series

```python
s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series([10, 20, 30, 40])
```

```python
# Suma
s1 + s2
```

```python
# Resta
s2 - s1
```

```python
# MultiplicaciÃ³n
s1 * s2
```

```python
# DivisiÃ³n
s1 / s2
```

```python
# Exponente
s1**2
```

```python
# Cuando las series no contiene la misma cantidad de elementos (Not a Number)
s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series([10, 20, 30])

s1 + s2
```
