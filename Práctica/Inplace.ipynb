# Inplace parameter

# if in place is True:
### It modifies an existing dataframe, returns nothing.
### `df.SomeOperation(Inplace = True)`

# If in place is False:
### It returns a copy of dataframe object with the performed operation(s), without modifying the existing dataframe.
### `df.SomeOperation(Inplace = False)`

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/marks_a.csv', usecols = [1,2,3,4])
data
```

```python
# Eliminamos una columna
df1 = data.drop('Grades', axis = 1)

# NOTA: Inplace es False por default
```

```python
# El dataframe original está intacto
data
```

```python
# La copia sufrió los cambios
df1
```

```python
data = pd.read_csv('/content/marks_a.csv', usecols = [1,2,3,4])
data

# Ahora modificaremos el dataframe original con inplace
data.drop('Grades', axis = 1, inplace = True)
# Ahora el dataframe original sí cambió
data
```
