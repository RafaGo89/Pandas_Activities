# nrows

```python
import pandas as pd
```

```python
# nrows permite leer una cantidad dada de filas del archivo al dataframe
data = pd.read_csv('/content/marks.csv', nrows = 3)
```

```python
data
```

# usecols

```python
# Usecols permite elegir que columnas usar al cargar el dataframe
data = pd.read_csv('/content/marks.csv', usecols = ['Name', 'Grades'])
data
```

```python
# También podemos pasar los índices de las columnas que queramos
data = pd.read_csv('/content/marks.csv', usecols = [0, 2])
data
```

```python
# También podemos pasar una lista guardada en una variable
list1 = ['Name', 'Marks']
data = pd.read_csv('/content/marks.csv', usecols = list1)
data
```

```python
# Leemos un archivo distinto
data = pd.read_csv('marks_new.csv')
data
```

```python
# Con esto transformamos las columnas de nuestro dataframe en minusculas
# y las hacemos coincidir con un lista de valores
data = pd.read_csv('marks_new.csv',
                   usecols = lambda x:x.lower() in ['aaa', 'bbb', 'ccc'])
data
```
