# dtype

```python
import pandas as pd
```

```python
# Con dtype podemos cambiar el tipo de dato de una columna
data = pd.read_csv('marks.csv', dtype = {'Marks':'float32'})
data
```

```python
# También podemos cambiar el tipo de dato de más de una columna a la vez
data = pd.read_csv('marks_new.csv', dtype = {'BBB':'float32', 'EEE':'float64'})
data
```

## true_values and false_values

```python
# Creación de archiv ya que no estaba en el repositorio
data = pd.read_csv('marks.csv')
# Añadimos una columna
data['Attended'] = ['Yes', 'Yes', 'Not', 'Yes', 'No', 'Yes', 'No']
data.to_csv('marks_a.csv')
```

```python
# Si usamos true_values y false_values, podemos transformar valores
# a True y Falso de acuerdo a una lista
data = pd.read_csv('marks_a.csv',
                   usecols = ['Name', 'Marks', 'Grades', 'Attended'],
                   true_values = ['Yes'],
                   false_values = ['No', 'Not'])
data
```

```python
# Encontrar cuantos estudiantes no asistieron
len(data)-data['Attended'].sum()
```
