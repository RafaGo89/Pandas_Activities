# Read a large CSV file in chunks

```python
import pandas as pd
```

```python
# Partimos el archivo en 10 chunks o pedazos
data = pd.read_csv('iris_code.csv', chunksize = 10)
data
```

```python
# Mostramos como está dividida la información del archivo
for chunk in data:
  print(chunk)
```

```python
# Concatenamos los chunks que tenemos para obtener el data frame completo
df = pd.concat(data, ignore_index = True)
df
```
