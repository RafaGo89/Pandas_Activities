# Pandas series can be created using
# List
# Tuple
# Dictionary
# Numpy Array
# Scalar value

### Pandas series (using list)

```python
import pandas as pd #Importamos la librería pandas
list1 = [11,22,33,"Hola",55.7]
x = pd.Series(list1) # Transformamos la lista en una serie de pandas
x

# x = pd.Series([11,22,33,"Hola",55.7]) Método directo
```

### Pandas series (using tuples)

```python
tuple1 = (33,4,90,8)
x = pd.Series(tuple1)
x

# x = pd.Series((33,4,90,8)) Método directo
```

### Pandas series (using dictionary)

```python
dict1 = {1:90, 2:12, 3:19, 10:11}
x = pd.Series(dict1)
x

# x = pd.Series({1:90, 2:12, 3:19, 10:11}) Método directo
```

### Pandas series (Using Numpy)

```python
import numpy as np #Importamos la librería Numpy
numpy1 = np.array([11,22,33,44,55])
x = pd.Series(numpy1)
x

# x = pd.Series(np.array([11,22,33,44,55])) Método directo
```

### Pandas series (Using Scalar value)

```python
x = pd.Series(12, index = [1,2,3,4,5])
x
```
