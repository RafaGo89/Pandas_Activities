## Creating a CSV File from a dataframe

```python
import pandas as pd
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
                'Grades':['AA','AB','AA','AB','AC','BA','BB']
               }
df1=pd.DataFrame(dict1)
df1
```

```python
# Transformamos el dataframe en un archivo CSV
df1.to_csv('Students.csv')
```
