# Viewing and inspecting dataframes

```python
import pandas as pd
```

```python
dict1 = {'Name': ['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks': [99,89,99,87,99,87,82]}

df = pd.DataFrame(dict1,index = ['stu1','stu2','stu3','stu4','stu5',
                             'stu6','stu7'])

df
```

## shape

```python
# Muestra el número de filas y columnas del dataframe
df.shape
```

```python
# Los valores que regresa se pueden almacenar
rows, columns = df.shape
print (rows, columns)
```

## head()

```python
# Muestra las 5 primeras filas del dataframe
df.head()
```

```python
# Podemos cambiar el número de filas que muestra
df.head(3)
```

## tail()

```python
# Muestra las últimas 5 filas del dataframe
df.tail()
```

```python
# Podemos cambiar el número de filas que muestra
df.tail(3)
```

## columns

```python
# Muestra el nombre de las columnas del dataframe
df.columns
```

## info()

```python
# Muestra un resumen de información básica del dataframe
df.info()
```

## Accesing to columns

```python
# Muestra la columna que escribamos
df['Name']
```

```python
# También existe este método con punto
df.Name
```

```python
df.Marks
```

```python
# Podemos mostrar múltiples columnas a la vez
df[['Name', 'Marks']]
```

## min()

```python
# El método min nos regresa el valor más pequeño de una columna
df['Marks'].min()
```

```python
df['Name'].min()
```

## max()

```python
# El método max regresa el valor más grande de una columna
df['Marks'].max()
```

## mean()

```python
# El método mean regresa la media aritmetica
df['Marks'].mean()
```

## describe()

```python
# Muestra información estadística básica acerca de la columna
df["Marks"].describe()
```

## value_counts()

```python
# Realiza un conteo de los valores únicos de la columna
df['Marks'].value_counts()
```

## count()

```python
# Realiza un conteo de valores no vacías en las columnas
df.count()
```

```python
df['Marks'].count()
```
