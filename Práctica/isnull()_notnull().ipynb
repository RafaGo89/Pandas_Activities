# Handling missing values

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/data_#na.csv')
data
```

## isnull()

```python
# Calcular total de valores nulos por columna en el dataframe
data.isnull().sum()
```

```python
# Calcular total de valores nulos en el dataframe
data.isnull().sum().sum()
```

```python
# isnal() hace lo mismo que isnull()
data.isna().sum()
```

```python
# Se pueden obtener los valores nulos por una sola columna o serie
data['Marks'].isnull().sum()
```

```python
# Podemos hacer consultas de los valores nulos
data[data['Marks'].isnull()]
```

## notnull()

```python
# Con notnull obtenemos los valores No Nulos por columna del dataframe
data.notnull().sum()
```

```python
# Con notnull obtenemos los valores No Nulos del dataframe
data.notnull().sum().sum()
```
