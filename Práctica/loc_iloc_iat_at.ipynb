# Diferencias entre **loc**, **iloc**, **iat** y **at**

```python
import pandas as pd
```

```python
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
                'Grades':['AA','AB','AA','AB','AC','BA','BB']
               }
df1=pd.DataFrame(dict1,index=['stu1','stu2','stu3','stu4','stu5',
                             'stu6','stu7'])

df1
```

## loc

```python
# Mostrar nombre, calificaciones y notas del estudiante 1 y estudiante 2
df1.loc[['stu1', 'stu2'], :]
```

```python
# Mostrar nombre y notas del estudiante 1, 3 y 5
df1.loc[['stu1', 'stu3', 'stu5'], ['Name', 'Grades']]
```

```python
# Mostrar nombre, calificaciones y notas desde el estudiante 1 al estudiante 7
df1.loc['stu1':'stu7', :]
```

## Iloc

```python
# Mostrar nombre, calificaciones y notas del estudiante 1 y 2
df1.iloc[[0, 1], :]
```

```python
# Mostrar nombre y notas del estudiante 1, 3 y 5
df1.iloc[[0, 2, 4], [0, 2]]
```

```python
# Mostrar nombre, calificaciones y notas desde el estudiante 1 al estudiante 7
df1.iloc[0:7, :]
```

# iat
##  Access a single value for a row/column pair by integer position,Similar to `iloc`, in that both provide integer-based lookups. Use `iat` if you only need to get or set a `single` value in a DataFrame or Series.

```python
# Mostrar nota del estudiante 2
df1.iat[1, 2]
```

```python
# Mostrar nombre del estudiante 5
df1.iat[4, 0]
```

```python
# Establecer la calificación del estudiante 5 a 99
df1.iat[4, 1] = 99
df1
```

# at
## Access a single value for a row/column label pair. Similar to `loc`, in that both provide label-based lookups. Use `at` if you only need to get or set a `single` value in a DataFrame or Series.

```python
# Mostrar nota del estudiante 2
df1.at['stu2', 'Grades']
```

```python
# Mostrar nombre del estudiante 5
df1.at['stu5', 'Name']
```

```python
# Establecer la calificación del estudiante 5 a 97
df1.at['stu5', 'Marks'] = 97
df1
```
