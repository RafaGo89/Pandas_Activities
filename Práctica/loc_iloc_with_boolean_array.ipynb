# Boolean array
## Boolean array size must be the same as your DataFrame Size.

```python
import pandas as pd
```

```python
import pandas as pd
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv'],
                'Marks':[98,89,99,87,90],
                'Grades':['AA','AB','AA','AB','AC']
               }
df1=pd.DataFrame(dict1,index=['stu1','stu2','stu3','stu4','stu5'])
df1
```

```python
# Creamos primero una lista booleana de la misma longitud que la cantidad de filas
# que tenga nuestro dataframe
df1.loc[[False, False, False, False, False]]
```

```python
# Ahora mismo, las 'banderas' est√°n apagadas, en falso
# Si las cambiamos a verdaderas, las encendemos, podemos comenzar a ver las filas
df1.loc[[True, False, False, False, False]]
```

```python
df1.loc[[True, False, False, True, False]]
```

```python
# Podemos hacer lo mismo si inlcuimos las filas y las columnas
# Filas y columnas tienen que tener la misma longitud que en nuestro dataframe
df1.loc[[False, False, False, False, False], [False, False, False]]
```

```python
df1.loc[[True, True, False, False, True], [True, False, True]]
```

```python
# Es indistinto si usamos loc o iloc
df1.iloc[[False, False, True, True, False], [False, True, True]]
```

# Ejemplos

```python
# Mostrar nombre del estudiante 1 y estudiante 2
df1.iloc[[True, True, False, False, False], [True, False, False]]
```

```python
# Mostrar notas del estudiante 1 y estudiante 5
df1.loc[[True, False, False, False, True], [False, False, True]]
```

```python
# Mostrar calificaci√≥n del estuadiante 4
df1.iloc[[False, False, False, True, False], [False, True, False]]
```
