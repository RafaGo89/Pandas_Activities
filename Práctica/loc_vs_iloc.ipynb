# loc vs iloc

```python
import pandas as pd
```

```python
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
                'Grades':['AA','AB','AA','AB','AC','BA','BB']
               }
df1=pd.DataFrame(dict1)
df1
```

# **loc**
## Access a group of rows and columns by label(s)
## dataframe.loc[row, column]
## **row is inclusive, columan is inclusive**
## it is inclusive

```python
# Mostrar la primera fila (índice 0) y todas las columnas
# : significa todas
df1.loc[0,:]
```

```python
# Mostrar fial 0, 1 y 2 con todas las columnas
df1.loc[[0, 1, 2], :]
```

```python
# Para lo anterior también se puede hacer esto
df1.loc[0:2, :]

# Al poner 0:2, comprende la fila 0, fila 1 y fila 2, eso significa inclusivo en loc
```

```python
# Mostrar todas las filas y la columna "Name"
df1.loc[:, 'Name']
```

```python
#  Mostrar todas las filas y las columnas "Name" y "Marks"
df1.loc[:, ['Name', 'Marks']]
```

```python
# Mostrar todas las filas, y las columnas desde 'Name'
df1.loc[:, 'Name':]
```

```python
df1.loc[:, 'Marks':]
```

# **iloc**
## Access a group of rows and columns by integer position(s)
## dataframe.iloc[row, column]
## **row is inclusive, columan is exclusive**
## It is exclusive

```python
# Mostrar todas las columnas de la primera columna
df1.iloc[:, 0]
```

```python
# Mostrar todas las filas de la columan 0 y 1
df1.iloc[:, 0:2]

# Como iloc es exclusivo, al poner 0:2, se incluye columna 0 y columna 1
# Es como si dijeramos, 2 - 1 = 1, columna 0 a la 1
```

# Second dataframe

```python
dict2 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
                'Grades':['AA','AB','AA','AB','AC','BA','BB']
               }
df2 = pd.DataFrame(dict1,index = ['stu1','stu2','stu3','stu4','stu5',
                             'stu6','stu7'])
df2
```

```python
# Mostrar nombre, calificaciones y nota del estudiante 1 y estudiante 2 (usando loc)
df2.loc[['stu1', 'stu2'], :]
```

```python
# Mostrar nombre del estudiante 1 al estudiante 4 (usando loc)
df2.loc['stu1':'stu4', 'Name']
```

```python
# Mostrar nombre y calificaciones del estudiante 1 y estudiante 2 (usando iloc)
df2.iloc[[0, 1], 0:2]
```

```python
# Mostrar nombre, calificaciones y notas del estudiante 1, 2 y 3 (usando iloc)
df2.iloc[[0, 1, 2], :]
```
