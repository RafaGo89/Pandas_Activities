# Handling missing values

```python
import pandas as pd
```

```python
data = pd.read_csv('data_m1.csv')
data
```

```python
# Con na_values, establecemos los valores que también se consideraran
# como valores NaN
data = pd.read_csv('data_m1.csv', na_values = ['not available', 'wrong data'])
data
```

```python
# Podemos aplicar na_values en una sola columna
ata = pd.read_csv('data_m1.csv',
                   na_values = {'Grades':['not available', 'wrong data']})
data
```
