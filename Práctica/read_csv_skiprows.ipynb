# skiprows

```python
import pandas as pd
```

```python
# Importamos el data frame del archvio CSV
data = pd.read_csv('/content/marks.csv')
data
```

```python
# Con skiprow quitamos filas al momento de leer el archivo
#Aquí quitamos la primera fila (el indice se acomoda automaticamente)
data = pd.read_csv('/content/marks.csv', skiprows = 1)
data
```

```python
#Aquí saltamos las 3 primeras filas
data = pd.read_csv('/content/marks.csv', skiprows = 3)
data
```

```python
#Aquí saltamos las filas 1, 3 y 5
data = pd.read_csv('/content/marks.csv', skiprows = [1, 3, 5])
data
```

```python
#Aquí quitamos un rango de filas, de la 1 a la 2, con un ciclo for
# Se debe de inciar en 1 porque si no quitariamos la fila de los encabezados
data = pd.read_csv('/content/marks.csv', skiprows = [i for i in range(1, 3)])
data
```

```python
# Aquí saltaríamos solo las filas 1, 3 y 5
data = pd.read_csv('/content/marks.csv', skiprows = [i for i in [1, 3, 5]])
data
```

```python
# Función para saber si un número es divisible entre 3
def function (x):
    if x%3 == 0:
      return True
    else:
      return False
```

```python
# Aquí saltaríamos solo las filas que sean divisibles entre 3
data = pd.read_csv('/content/marks.csv', skiprows = lambda x: function(x))
data
```

```python
# Aquí saltaríamos las 3 últimas filas del dataframe
data = pd.read_csv('/content/marks.csv', skipfooter = 3, engine = 'python')
data
```
